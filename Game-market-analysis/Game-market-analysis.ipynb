# Интернет-магазин "Стримчик"

Из открытых источников доступны исторические данные до 2016 г о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). 

**Цель исследования** — нужно выявить определяющие успешность игры закономерности:
1. Нужно отработать принцип работы с данными для прогноза на следующий год имея данные предыдущего года
2. Чтобы сделать ставку на потенциально популярный продукт и спланировать рекламные кампании;
3. Составьте портрет пользователя каждого региона
4. Проверить гипотезы
    1.  Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
    2. Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.


**Ход исследования**

Данные собраны в файл `games.csv`. О качестве данных ничего не известно. Поэтому перед проверкой гипотез понадобится обзор данных. 

Проверим данные на ошибки и оценим их влияние на исследование. Затем, на этапе предобработки поищем возможность исправить самые критичные ошибки данных.
 
Таким образом, исследование пройдёт в три этапа:
 1. Обзор данных.
 2. Предобработка данных.
 3. Проверка гипотез.

В рейтинге пользователей есть аббревиатура `tbd` означающая **To be determined - будет определено** (в смысле "будет определено позже")
Можно все пропуски заменить на `tbd` это один вариант, либо разделить все игры по категериям и в каждой категории наити медианное значение по рейтингу и заменить пропуски. Затем перевести все `tbd` в **0 - оценки нет** и столбец перевести в float чтобы можно было корреляцию посмотреть. 

Анализ показал, что замена по предыдущему предположению нельзя назвать это хорошим решением, так как оно заняло достаточно времени, но важнее всего то, что отсутствие оценок критиков/пользователей абсолютно нормальная ситуация. Поэтому заполнение пропусков может исказить данные в этом случае. В связи с этим лучше оставить эти пустые значения как есть и tbd также сделать пустыми значениями.**Поэтому заменили tbd на пустые значения**

```python
display(games['critic_score'].unique())# проверка оценки критиков
display(games['critic_score'].value_counts())# подсчет количества по оценки критиков
```

```python
display(games['platform'].unique())# проверка платформы
display(games['platform'].value_counts())# подсчет количества по платформе
```

```python
display(games['year_of_release'].unique())# проверка по году выпуска
display(games['year_of_release'].value_counts())# подсчет количества по году выпуска
```

```python
display(len(games[games['critic_score'].isna()]))
display(games)
```

Заменяем tbd на пустые значения и переводим в float

```python
games.loc[games['user_score'] == 'tbd', 'user_score'] = np.nan
display(games['user_score'].value_counts())        
```

```python
games['user_score']=(games['user_score'].astype('float'))
```

```python
games.info()
```

```python
# Замени все пропуски 'rating' на -1
games['rating'] = games['rating'].fillna(-1)
display(games.loc[games['rating']==-1].head(10))
display(games)
```

```python
for i in range(len(games)):
    if (games.loc[i,'rating'] ==-1):
        games.loc[i,'rating']='no_rating'
```

```python
display(games)
```

```python
# display(games.corr())#смотрим корреляцию
```

 Суммируем все продажи по регионам и записываем в столбец `all_sales`

```python
for i in range(len(games)):
    games.loc[i,'all_sales']=games.loc[i,'na_sales']+games.loc[i,'eu_sales']+games.loc[i,'jp_sales']+games.loc[i,'other_sales']
```

## Анализ данных

### Анализ игр и платформ

Смотрим сколько игр выпускалось в разные годы и сколько суммарно продаж

```python
games_year_group=games.pivot_table(index='year_of_release',values='all_sales', aggfunc=['sum','count'])
games_year_group.columns = ['all_sales', 'count']
display(games_year_group.sort_values(by='all_sales',ascending = False))
```

```python
games_year_group.sort_values(by='year_of_release',ascending = True).plot(y='all_sales',grid=True)
plt.title('Выпуск игр по годам')
plt.show()
```

```python
platform_year_group=games.pivot_table(index=['platform'],values='all_sales', aggfunc=['sum','count'])
platform_year_group.columns = ['all_sales', 'count']
display(platform_year_group.sort_values(by='all_sales',ascending = False))
# display(games.pivot_table(index=['year_of_release'],values='all_sales', aggfunc=['sum']))
```

```python
platform_year_group.sort_values(by='all_sales',ascending = False).plot(y='all_sales',kind='bar',grid=True)
plt.title('Количество продаж по платформам')
plt.show()
```

```python
platform_year_group.sort_values(by='count',ascending = False).plot(y='count',kind='bar',grid=True)
plt.title('Количество игр по платформам')
plt.show()
```

## Проверка гипотез

1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
2. Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

**Средние пользовательские рейтинги платформ Xbox One и PC одинаковые**

Нулевая гипотеза - Проверим гипотезу о равенстве пользовательских рейтингов платформ Xbox One и PC 

Альтернативная гипотеза - Пользовательские рейтинги платформ Xbox One и PC разные

В данных есть пропуски поэтому их удалим

```python

display(platform_xone)
xone=platform_xone.loc[:,'user_score']
display(xone)
```

```python
display(xone.isna().sum()) # подсчёт пропусков
```

```python
xone=xone.dropna()
display(xone)
```

```python
platform_pc=games.query('platform== "PC" and year_of_release in ("2014", "2015", "2016")')
display(platform_pc)
pc=platform_pc.loc[:,'user_score']
display(pc)
display(pc.isna().sum()) # подсчёт пропусков
```

```python
pc=pc.dropna()
display(pc)
```

```python
alpha = 0.01# критический уровень статистической значимости

#проверьте гипотезу о равенстве средних двух независимых совокупностей
results = st.ttest_ind(
    xone, 
    pc,
    equal_var = False)# выборки большие но не равные поэтому поставили equal_var = False

print('p-значение:',results.pvalue )

if results.pvalue < alpha: # сравните получившееся p-значение с заданным уровнем статистической значимости:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Значит, можно сделать вывод, что пользовательские рейтинги одинаковы для платформ Xbox One и PC
